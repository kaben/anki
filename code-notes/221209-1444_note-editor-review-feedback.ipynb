### How search works:

- Python code calls `Collection.find_cards(query, order, reverse)`
- `Collection.find_cards()` calls `Collection._build_sort_mode(order, reverse, 'C')` to construct a `mode` instance of type `SortOrder`.
    - `order` can bee a `bool`, `str`, or `BroswerColumns.Column`.
        - If `bool` and `False`, no ordering.
        - If `bool` and `True`, looks like a sort column is looked up in the config table of the database. The config key depends on the browser mode:
            - For Reviews mode, it's `reviewSortType`.
            - For Cards mode, it's `sortType`.
            - For Notes mode, it's `noteSortType`.
            - The sort column is then looked up via `Collection.get_browser_column`.
                - This in turns looks it up via backend method `all_browser_columns`, which returns a list of `enum Column` defined in `rslib/src/browser_table.rs`.
        - Looks like a custom order can also be specified. It may be that the sort order is then determined by a setting in the "Fields" preferences for the note type, where one of the fields can have the option "Sort by this field in the browser" checked.
- `Collection.find_cards()` then calls `Collection._backend.search_cards(search=query, order=mode`), which is defined in `proto/anki/search.proto` and implemented in `rslib/src/backend/search/mod.rs` as `fn SearchService.search_cards`.
- `fn SearchService.search_cards()` forwards to `fn Collection.search_cards<N>(search: N, mode: SortMode)` where `N` is of type `trait TryIntoSearch`.
    - `TryIntoSearch` provides implementations of `fn try_into_search()` for types `str`, `String`, and types that can convert to `str` or `String` via `into()`.
- `fn Collection.search_cards<N>()` forwards to `fn Collection.search<T, N>(search: N, mode: SortMode)` where `T` is of type `FromSql + AsReturnType`.
    - `FromSql` is from `crate rusqlite::types`, and is a trait for types that can be create from the results of Sqlite queries.
    - `AsReturnItemType` implements `fn as_return_item_type() -> ReturnitemType` for each of `CardId`, `NoteId`, `RevlogId`, which respectively return `ReturnItemType::Cards`, `ReturnItemType::Notes`, or `ReturnItemType::Reviews`.
    - `ReturnItemType` provides `fn ReturnItemType.required_table() -> RequiredTable` for each of of the above `ReturnItemType::` enums, which respectively return `RequiredTable::Reviews`, `RequiredTable::Cards`, or `RequiredTable::Notes`.
- `fn Collection.search<T, N>()` then:
    - Calls `T::as_return_item_type()` to obtain `ReturnItemType::Cards` (because we're seaching for card IDs).
    - Calls `search.try_into_search()` to obtain a vector of `enum search::parser::Node`. There are five types:
        - `And`
        - `Or`
        - `Not(Box<Node>)`
        - `Group(Vec<Node>)`
        - `Search(SearchNode)`.
            - `enum search::parser::SearchNode` has types for each of the search keys that can be searched:
                - "deck"
                - "note"
                - "card"
                - "review"
                - "tag"
                - ...
    - `try_into_search()` in turn calls `fn seach::parser::parse(input: &str) -> Result<Vec<Node>>`
