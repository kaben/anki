```python
def generate_problems_for(section_name, section_tags, chapter, section, num_problems, note_type_dict, base_tags):
    did = mw.col.decks.id_for_name(section_name)
    if did is None:
        deck = mw.col.decks.add_normal_deck_with_name(section_name)
        did = mw.col.decks.id_for_name(section_name)
    for i in range(1, num_problems + 1):
        note = mw.col.new_note(note_type_dict)
        note['Front'] = note_name_fmt.format(c=chapter, s=section, n=i)
        note.tags.extend(base_tags)
        note.tags.extend(section_tags)
        print(did, note['Front'])
        mw.col.add_note(note, did)    
```

```python
import aqt
browser = aqt.dialogs._dialogs['Browser'][1]
table = browser.table
model = table._model
cards = [model._state.get_card(item) for item in model._items]
nt = cards[0].note().note_type()
#nt
```

```python
for (section_name, section_tags, chapter, section, num_problems) in section_data:
    generate_problems_for(section_name, section_tags, chapter, section, num_problems, nt, base_tags)
```
