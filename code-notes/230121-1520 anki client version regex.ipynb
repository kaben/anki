```python
import re
```

```python
client_version_re = re.compile(r"""
^
(?P<name>\S+)
,
(?P<majorver>0|[1-9]\d*)
\.
(?P<minorver>0|[1-9]\d*)
\.
(?P<patchver>0|[1-9]\d*)
(?:-(?P<prerelease>
    (?:0|[1-9]\d*|\d*[a-zA-Z-][0-9a-zA-Z-]*)
    (?:\.(?:0|[1-9]\d*|\d*[a-zA-Z-][0-9a-zA-Z-]*))*
  )
)?(?:\+(?P<buildmetadata>
    [0-9a-zA-Z-]+(?:\.[0-9a-zA-Z-]+)*
  )
)?
\s+
\(
(?P<buildhash>[0-9a-zA-Z-]+)
\.?
(?P<variant>[0-9a-zA-Z-]+(?:\.[0-9a-zA-Z-]+)*)?
\)
,
(?P<platform>\S+)
$
""", re.VERBOSE)
client_version_re
```

```python
short_client_version_re = re.compile(r"""
^
(?P<majorver>0|[1-9]\d*)
\.
(?P<minorver>0|[1-9]\d*)
\.
(?P<patchver>0|[1-9]\d*)
(?:-(?P<prerelease>
    (?:0|[1-9]\d*|\d*[a-zA-Z-][0-9a-zA-Z-]*)
    (?:\.(?:0|[1-9]\d*|\d*[a-zA-Z-][0-9a-zA-Z-]*))*
  )
)?(?:\+(?P<buildmetadata>
    [0-9a-zA-Z-]+(?:\.[0-9a-zA-Z-]+)*
  )
)?
,
(?P<buildhash>[0-9a-zA-Z-]+)
\.?
(?P<variant>[0-9a-zA-Z-]+(?:\.[0-9a-zA-Z-]+)*)?
,
(?P<platform>\S+)
$
""", re.VERBOSE)
client_version_re
```

```python
client_version_re = re.compile(r"^(?P<name>\S+),(?P<majorver>0|[1-9]\d*)\.(?P<minorver>0|[1-9]\d*)\.(?P<patchver>0|[1-9]\d*)(?:-(?P<prerelease>(?:0|[1-9]\d*|\d*[a-zA-Z-][0-9a-zA-Z-]*)(?:\.(?:0|[1-9]\d*|\d*[a-zA-Z-][0-9a-zA-Z-]*))*))?(?:\+(?P<buildmetadata>[0-9a-zA-Z-]+(?:\.[0-9a-zA-Z-]+)*))?\s+\((?P<buildhash>[0-9a-zA-Z-]+)\.?(?P<variant>[0-9a-zA-Z-]+(?:\.[0-9a-zA-Z-]+)*)?\),(?P<platform>\S+)$")
client_version_re
```

```python
anki_client_version_string = "anki,2.1.57 (6ad33e70.mxyzptlk),lin:ubuntu:22.04"
client_version_match = client_version_re.match(anki_client_version_string)
client_version_match.groupdict() if client_version_match else ""
```

```python
semver_re = re.compile(r"""
^
(?P<major>0|[1-9]\d*)
\.
(?P<minor>0|[1-9]\d*)
\.
(?P<patch>0|[1-9]\d*)
(?:-(?P<prerelease>
    (?:0|[1-9]\d*|\d*[a-zA-Z-][0-9a-zA-Z-]*)
    (?:\.(?:0|[1-9]\d*|\d*[a-zA-Z-][0-9a-zA-Z-]*))*
  )
)?(?:\+(?P<buildmetadata>
    [0-9a-zA-Z-]+(?:\.[0-9a-zA-Z-]+)*
  )
)?$
""", re.VERBOSE)
semver_re
```

```python
semver_match = semver_re.match('1.0.0-beta+exp.sha.5114f85')
semver_match.groupdict()
```
